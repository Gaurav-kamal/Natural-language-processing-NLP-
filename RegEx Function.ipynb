```python
import re
```

#search() function

```python
txt = 'The rain in Spain'
x = re.search('^The.*Spain$', txt)
print(x)
```

```python
txt = 'The rain in Spain'
x = re.search('Portugal', txt)
print(x)
```

#findall() function

```python
txt = 'The rain in Spain'
x = re.findall('ai', txt)
print(x)
```

```python
txt = 'The rain in Spain'
x = re.findall('Portugal', txt)
print(x)
```

```python
txt = 'The rain in Spain'
x = re.search('\s', txt)
print('The first white.space character is located in position:', x.start())
```

#split() function

```python
txt = 'The rain in Spain'
x = re.split("\s", txt)
print(x)
```

```python
txt = 'The rain in Spain'
x = re.split("\s", txt, 1)
print(x)
```

#sub() function

```python
txt = 'The rain in Spain'
x = re.sub("\s", '9', txt)
print(x)
```

```python
txt = 'The rain in Spain'
x = re.sub("\s", '9', txt, 2)
print(x)
```

```python
txt = 'The rain in Spain'
x = re.search(r'\bS\w+', txt)
print(x.span())
```

```python
txt = 'The rain in Spain'
x = re.search(r'\bS\w+', txt)
print(x.string)
```

```python
txt = 'The rain in Spain'
x = re.search(r'\bS\w+', txt)
print(x.group())
```
